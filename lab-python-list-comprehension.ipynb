````python

products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {articulo: int(input(f"Introduce la cantidad de {articulo}s disponible: ")) for articulo in products}
    return inventory

def get_customers_orders():
    num_orders = int(input("Enter the number of customer orders: "))
    customers_orders = {input(f"Introduce el nombre del artículo que quiere ({i + 1}/{num_orders}): ").strip().lower() for i in range(num_orders)}
    return customers_orders

def calculate_total_price(customers_orders):
    prices = {articulo: float(input(f"Introduce el precio de {articulo}: ")) for articulo in customers_orders}
    total_price = sum(prices.values())
    return total_price

def update_inventory(customers_orders, inventory):
    inventory = {
        articulo: (cantidad - 1 if articulo in customers_orders and cantidad > 0 else cantidad)
        for articulo, cantidad in inventory.items()
    }
    
    inventory = {articulo: cantidad for articulo, cantidad in inventory.items() if cantidad > 0}
    return inventory

def print_order_statistics(order_statistics):
    total_products_ordered, percentage_ordered = order_statistics
    print("Order Statistics:\n")
    print(f"Total products ordered: {total_products_ordered}")
    print(f"Percentage of unique products ordered: {percentage_ordered}%")

def calculate_order_statistics(customers_orders, products):
    total_products_ordered = len(customers_orders)
    total_unique_products = len(products)
    percentage_ordered = (total_products_ordered / total_unique_products) * 100
    return total_products_ordered, percentage_ordered

def total_price(customers_orders):
    precio = {
        articulo: float(input(f"Introduce el precio del {articulo}: "))
        for articulo in customers_orders
    }
    precio_total = sum(precio.values())
    return precio_total, precio

def print_updated_inventory(inventory):
    print("Updated Inventory: \n")
    for articulo, cantidad in inventory.items():  
        print(f"{articulo}: {cantidad}")

````

````python
def main():
    print("Inicializando inventario...")
    inventory = initialize_inventory(products)

    print("\nObteniendo los pedidos de los clientes...")
    customers_orders = get_customers_orders()

    print("\nCalculando el precio total de los pedidos...")
    total_price_value = calculate_total_price(customers_orders)
    print(f"\nPrecio total de todos los pedidos: ${total_price_value:.2f}")

    print("\nActualizando el inventario según los pedidos...")
    inventory = update_inventory(customers_orders, inventory)

    print("\nImprimiendo el inventario actualizado...")
    print_updated_inventory(inventory)

    print("\nCalculando las estadísticas de los pedidos...")
    order_statistics = calculate_order_statistics(customers_orders, products)
    print_order_statistics(order_statistics)
    
    print("\nCalculando el precio total nuevamente (si es necesario)...")
    total_price_value, item_prices = total_price(customers_orders)
    print(f"\nPrecios detallados: {item_prices}")
    print(f"Precio total de todos los pedidos: ${total_price_value:.2f}")
````

````python
main()
````
